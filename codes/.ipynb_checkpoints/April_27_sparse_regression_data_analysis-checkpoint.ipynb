```julia

## This function does the following: given two vectors x and y, it computes how many components of x and y differs significantly

function l_0_norm_diff(x,y,n,delta_l0_norm)
    diffxy = x-y
    l_0_norm_xy = ones(n)
    for i in 1:n
        if abs(diffxy[i]) <= delta_l0_norm
            l_0_norm_xy[i]=0
        end
    end
    return sum(l_0_norm_xy)
end

## checking cardinality
function card(x, n, delta_l0_norm)
    return l_0_norm_diff(x,zeros(n),n,delta_l0_norm)
end

## computes support of a signal
function supp_calc(x, tol_supp) # tol_supp is the tolerance for support recovery, we set it 1e-3 for our experiment
    supp_x = similar(x)
    for i in 1:length(supp_x)
        if abs(x[i]) > tol_supp
            supp_x[i] = 1
        else
            supp_x[i] = 0
        end
    end
    return supp_x
end

function exact_support_recovery(x_true, x_computed, tol_supp)
    # x_true: computed signal
    # x_computed : signal computed by our algorithm
    supp_true = supp_calc(x_true, tol_supp)
    n = length(supp_true)
    supp_computed = supp_calc(x_computed, tol_supp)
    exact_supp_recovery = 100*(n-l_0_norm_diff(supp_true, supp_computed, n, tol_supp))/n
    return exact_supp_recovery
end
```

```julia

length_m_array = length(m_array)
length_instance_m_array = length(instance_m_array)
array_fixed_point_gap = zeros(length_m_array, length_instance_m_array)
array_absolute_optimality_gap = zeros(length_m_array, length_instance_m_array)
array_distance_from_optimal_sol = zeros(length_m_array, length_instance_m_array)
array_optimality_gap_relative_to_signal = zeros(length_m_array, length_instance_m_array)
array_optimality_gap_relative_to_lasso = zeros(length_m_array, length_instance_m_array)
array_dist_solPRS_sparse_set = zeros(length_m_array, length_instance_m_array)
array_support_recovery = zeros(length_m_array, length_instance_m_array)
array_support_recovery_lasso = zeros(length_m_array, length_instance_m_array)
array_support_recovery_ratio = zeros(length_m_array, length_instance_m_array)

## time to load the data from the dictionaries into the arrays

delta_l0_norm = 0.05
tol_supp = 1e-3

for i_m in 1:length_m_array
    m = m_array[i_m][1]
    n = 2*m
    for i_instance_m in 1:length_instance_m_array

        # input corresponding size m Ã— n, and random instance number
        A, b, k, v, m, n, c, d, xhat = dict_sparse_reg[(m, i_instance_m)]

        # output corresponding to that instance
        best_sol_x, best_sol_y, best_sol_z, fixed_point_gap, obj_value_sol_PRS, dist_solPRS_sparse_set, best_objective_gurobi, best_solution_x_gurobi, best_solution_y_gurobi, best_objective_lasso, best_solution_lasso = dict_output_sparse_reg[(m, i_instance_m)]


        # time to calculate the objective value of the original signal
        obj_value_sol_PRS = norm(A*best_sol_x - b)^2
        objective_val_signal = norm(A*xhat-b)^2
        support_rec_val = exact_support_recovery(xhat, best_sol_x, tol_supp)# l_0_norm_diff(best_sol_y, xhat, n, delta_l0_norm)
        support_rec_val_lasso = exact_support_recovery(xhat, best_solution_lasso, tol_supp)# l_0_norm_diff(best_sol_y, xhat, n, delta_l0_norm)
        support_rec_val_ratio = support_rec_val/support_rec_val_lasso
        
        # time to create all the values

        array_fixed_point_gap[i_m, i_instance_m] = fixed_point_gap
        array_absolute_optimality_gap[i_m, i_instance_m] = best_objective_gurobi/obj_value_sol_PRS
        # array_distance_from_optimal_sol[i_m, i_instance_m] = norm(best_sol_y - best_solution_x_gurobi)/norm(best_solution_x_gurobi)
        array_optimality_gap_relative_to_signal[i_m, i_instance_m] = objective_val_signal/obj_value_sol_PRS
        array_optimality_gap_relative_to_lasso[i_m, i_instance_m] = best_objective_lasso/obj_value_sol_PRS
        array_dist_solPRS_sparse_set[i_m, i_instance_m] = dist_solPRS_sparse_set
        array_support_recovery[i_m, i_instance_m] = support_rec_val
        array_support_recovery_lasso[i_m, i_instance_m] = support_rec_val_lasso
        array_support_recovery_ratio[i_m, i_instance_m] = support_rec_val_ratio
        

        end
end
```
